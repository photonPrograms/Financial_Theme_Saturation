```python
def generate_archive_links(start_date, end_date):
    archive_links = []
    
    current_date = start_date
    while current_date <= end_date:
        year_month = current_date.strftime('%Y-%m')
        day = current_date.strftime('%d')
        archive_links.append(f"https://www.reuters.com/archive/{year_month}/{day}/1")
        current_date += timedelta(days=1)  # Move to the next day

    return archive_links

start_date = datetime(2006, 5, 22)
end_date = datetime(2023, 10, 31)

generated_links = generate_archive_links(start_date, end_date)
```
